```python
# ÂÆâË£ÖÂøÖË¶ÅÁöÑÂ∫ì
!pip install transformers torch

# ÂØºÂÖ•Â∫ì
from transformers import pipeline, set_seed

# Âä†ËΩΩÊñáÊú¨ÁîüÊàê pipeline
# Êàë‰ª¨‰ΩøÁî®‰∏Ä‰∏™ËæÉÂ∞èÁöÑÊ®°Âûã‰ª•‰æøÂú®Colab‰∏≠Âø´ÈÄüËøêË°å
generator = pipeline('text-generation', model='gpt2')
set_seed(42)  # ËÆæÁΩÆÈöèÊú∫ÁßçÂ≠êÔºå‰øùËØÅÁªìÊûúÂèØÂ§çÁé∞

# ÁÆÄÂçïÊñáÊú¨ÁîüÊàêÁ§∫‰æã
print("=== ÁÆÄÂçïÊñáÊú¨ÁîüÊàêÁ§∫‰æã ===")
prompt = "Artificial intelligence can"
results = generator(prompt, max_length=50, num_return_sequences=1)
print(results[0]['generated_text'])




```

```python
# ‰∫íÂä®ÊñáÊú¨ÁîüÊàê
print("\n=== ‰∫íÂä®ÊñáÊú¨ÁîüÊàê ===")
def interactive_advanced_generator():
    while True:
        prompt = input("\nËØ∑ËæìÂÖ•‰∏Ä‰∏™ÊñáÊú¨ÂºÄÂ§¥ÔºàËæìÂÖ•'quit'ÈÄÄÂá∫Ôºâ: ")
        if prompt.lower() == 'quit':
            break

        length = input("ËØ∑ËæìÂÖ•ÁîüÊàêÁöÑÊúÄÂ§ßÈïøÂ∫¶ÔºàÈªòËÆ§100Ôºâ: ")
        length = int(length) if length.isdigit() else 100

        num_samples = input("ËØ∑ËæìÂÖ•ÁîüÊàêÁöÑÊ†∑Êú¨Êï∞Ôºà1-3ÔºåÈªòËÆ§1Ôºâ: ")
        num_samples = int(num_samples) if num_samples.isdigit() and 1 <= int(num_samples) <=3 else 1

        print("\nÁîüÊàê‰∏≠...")
        results = generator(
            prompt,
            max_length=length,
            num_return_sequences=num_samples,
            temperature=0.7,  # ÊéßÂà∂ÈöèÊú∫ÊÄßÔºåÂÄºË∂äÂ∞èË∂äÁ°ÆÂÆö
            do_sample=True
        )

        for i, result in enumerate(results):
            print(f"\nÊ†∑Êú¨ {i+1}:")
            print(result['generated_text'])

# ËøêË°å‰∫íÂä®ÁîüÊàêÂô®
interactive_advanced_generator()
```

```python
# Â∞ùËØï‰∏çÂêåÁöÑÊ®°ÂûãÔºàÂèØÈÄâÔºâ
print("\n=== Â∞ùËØï‰∏çÂêåÁöÑÊ®°Âûã ===")
print("ËÆ©Êàë‰ª¨ËØïËØï‰∏Ä‰∏™‰∏ìÈó®Áî®‰∫éÈóÆÁ≠îÁöÑÊ®°Âûã")

# Âä†ËΩΩÈóÆÁ≠îÊ®°Âûã
question_answerer = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")

def interactive_qa():
    context = """
    ËØ≠Ë®ÄÊ®°ÂûãÊòØ‰∏ÄÁßç‰∫∫Â∑•Êô∫ËÉΩÁ≥ªÁªüÔºåËÉΩÂ§üÁêÜËß£ÂíåÁîüÊàê‰∫∫Á±ªËØ≠Ë®Ä„ÄÇ
    ÂÆÉ‰ª¨ÈÄöËøáÂàÜÊûêÂ§ßÈáèÊñáÊú¨Êï∞ÊçÆÊù•Â≠¶‰π†ËØ≠Ë®ÄÊ®°ÂºèÂíåËßÑÂæã„ÄÇ
    Áé∞‰ª£ËØ≠Ë®ÄÊ®°ÂûãÂ¶ÇGPT‰ΩøÁî®TransformerÊû∂ÊûÑÔºåËÉΩÂ§üÁîüÊàêËøûË¥ØÁöÑÊñáÊú¨Âπ∂ÁêÜËß£‰∏ä‰∏ãÊñá„ÄÇ
    Ëøô‰∫õÊ®°ÂûãÂú®ËÅäÂ§©Êú∫Âô®‰∫∫„ÄÅÊú∫Âô®ÁøªËØë„ÄÅÊñáÊú¨ÊëòË¶ÅÁ≠âÈ¢ÜÂüüÊúâÂπøÊ≥õÂ∫îÁî®„ÄÇ
    """

    print("\nÂ∑≤Áü•‰ø°ÊÅØ:")
    print(context)

    while True:
        question = input("\nËØ∑ËæìÂÖ•‰∏Ä‰∏™ÈóÆÈ¢òÔºàËæìÂÖ•'quit'ÈÄÄÂá∫Ôºâ: ")
        if question.lower() == 'quit':
            break

        result = question_answerer(question=question, context=context)
        print(f"Á≠îÊ°à: {result['answer']}")
        print(f"ÁΩÆ‰ø°Â∫¶: {result['score']:.4f}")

# ËøêË°åÈóÆÁ≠î‰∫íÂä®
interactive_qa()
```

```python
# ÂÆâË£ÖÂøÖË¶ÅÁöÑÂ∫ì
!pip install transformers torch gradio

# ÂØºÂÖ•ÊâÄÈúÄÂ∫ì
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM
import gradio as gr
import random

# Ê£ÄÊü•GPUÂèØÁî®ÊÄß
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"‰ΩøÁî®ËÆæÂ§á: {device}")

# Âä†ËΩΩ**Â§öËØ≠Ë®ÄÊîØÊåÅÁöÑ‰∏≠ÊñáÁîüÊàêÊ®°Âûã**ÔºàËß£ÂÜ≥ÂàÜËØçÂô®‰∏çÂÖºÂÆπÈóÆÈ¢òÔºâ
model_name = "bigscience/bloom-560m"  # ÊîØÊåÅÂ§öËØ≠Ë®ÄÔºåÂØπ‰∏≠ÊñáÁîüÊàêÂèãÂ•Ω

print("Ê≠£Âú®Âä†ËΩΩÊ®°ÂûãÂíåÂàÜËØçÂô®...")
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name).to(device)

# ËØóÊ≠åÁîüÊàêÂáΩÊï∞ÔºàÈÄÇÈÖçÊñ∞Ê®°ÂûãÔºâ
def generate_poem(prompt, style, length=100, creativity=0.7):
    """Ê†πÊçÆ‰∏≠ÊñáÊèêÁ§∫ÁîüÊàêÂØπÂ∫îÈ£éÊ†ºÁöÑ‰∏≠ÊñáËØóÊ≠å"""
    # ‰∏≠ÊñáÈ£éÊ†ºÂºïÂØºËØçÔºàÊòéÁ°ÆÁîüÊàêÊñπÂêëÔºâ
    style_guide = {
        "Áé∞‰ª£ËØó": "ÂÜô‰∏ÄÈ¶ñÁé∞‰ª£ËØóÔºåÊÑèË±°ÁÅµÂä®ÔºåÂè•ÂºèËá™Áî±Ôºö\n",
        "Âè§ËØó": "ÂÜô‰∏ÄÈ¶ñ‰∫îË®Ä/‰∏ÉË®ÄÂè§ËØóÔºåÊäºÈüµÂ∑•Êï¥ÔºåÊÑèÂ¢ÉÊÇ†ËøúÔºö\n",
        "Áà±ÊÉÖËØó": "ÂÜô‰∏ÄÈ¶ñÁà±ÊÉÖËØóÔºåÊÉÖÊÑüÁªÜËÖªÔºåËØ≠Ë®ÄÊ∏©ÊüîÔºö\n",
        "Ëá™ÁÑ∂ËØó": "ÂÜô‰∏ÄÈ¶ñËá™ÁÑ∂‰∏ªÈ¢òÁöÑËØóÔºåÊèèÁªòÂ±±Ê≤≥ËçâÊú®ÔºåÂÖÖÊª°ÁîüÊú∫Ôºö\n",
        "Âì≤ÁêÜËØó": "ÂÜô‰∏ÄÈ¶ñÂì≤ÁêÜËØóÔºåËï¥Âê´‰∫∫ÁîüÊÄùËÄÉÔºåÂºï‰∫∫Ê∑±ÊÄùÔºö\n"
    }.get(style, "ÂÜô‰∏ÄÈ¶ñËØóÔºö\n")

    # ÊûÑÂª∫ÂÆåÊï¥‰∏≠ÊñáÊèêÁ§∫
    full_prompt = f"{style_guide}'{prompt}'\n"

    # ÁºñÁ†Å‰∏≠ÊñáÊèêÁ§∫ÔºàËá™Âä®ÈÄÇÈÖçÊ®°ÂûãÂàÜËØçÂô®Ôºâ
    input_ids = tokenizer.encode(full_prompt, return_tensors="pt").to(device)
    input_length = len(input_ids[0])  # ËÆ∞ÂΩïËæìÂÖ•ÈïøÂ∫¶ÔºåÊéßÂà∂ÁîüÊàêÂ¢ûÈáè

    # ÁîüÊàêËØóÊ≠åÔºà‰ºòÂåñÂèÇÊï∞ÈÄÇÈÖç‰∏≠ÊñáÔºâ
    with torch.no_grad():
        output = model.generate(
            input_ids,
            max_length=input_length + length,  # ËæìÂÖ•ÈïøÂ∫¶ + ÊúüÊúõÁöÑÂ¢ûÈáèÈïøÂ∫¶
            num_return_sequences=1,
            temperature=creativity,          # ÂàõÈÄ†ÊÄßÔºà0.3-1.2ÔºåË∂äÈ´òË∂äÂ§©È©¨Ë°åÁ©∫Ôºâ
            top_k=30,                        # ÈááÊ†∑Êó∂‰øùÁïôÂâç30‰∏™È´òÊ¶ÇÁéáËØç
            top_p=0.9,                       # Á¥ØÁßØÊ¶ÇÁéáËææ0.9Êó∂ÂÅúÊ≠¢ÈááÊ†∑
            repetition_penalty=1.2,          # Èôç‰ΩéÈáçÂ§çË°®Ëø∞ÁöÑÊ¶ÇÁéá
            do_sample=True,                  # ÂêØÁî®ÈöèÊú∫ÈááÊ†∑ÔºàÂ¢ûÂä†Â§öÊ†∑ÊÄßÔºâ
            pad_token_id=tokenizer.eos_token_id,  # Â°´ÂÖÖÁî®ÁöÑtokenÔºàËá™Âä®ÂåπÈÖçÊ®°ÂûãÔºâ
        )

    # Ëß£Á†ÅÂπ∂ÊèêÂèñËØóÊ≠åÔºàÂéªÈô§ÊèêÁ§∫ÂâçÁºÄÔºâ
    generated_text = tokenizer.decode(output[0], skip_special_tokens=True)
    poem = generated_text[len(full_prompt):].strip()  # Âè™‰øùÁïôÁîüÊàêÁöÑËØóÊ≠åÈÉ®ÂàÜ

    # ÂÖúÂ∫ïÈÄªËæëÔºöËã•ÁîüÊàêÂÜÖÂÆπ‰∏∫Á©∫ÔºåËøîÂõûÊèêÁ§∫
    if not poem:
        return "Êä±Ê≠âÔºåÊöÇÊó∂Êú™ÁîüÊàêËØóÊ≠å~ ÂèØÂ∞ùËØïÊõ¥Êç¢‰∏ªÈ¢òÊàñË∞ÉÊï¥„ÄåÂàõÈÄ†ÊÄß„ÄçÂèÇÊï∞„ÄÇ"

    # Ê†ºÂºèÂåñËØóÊ≠åÔºàÊ∑ªÂä†Êç¢Ë°åÔºå‰ºòÂåñÊéíÁâàÔºâ
    return format_poem(poem)

def format_poem(text):
    """Áªô‰∏≠ÊñáËØóÊ≠åÊ∑ªÂä†Êç¢Ë°åÔºåÊ®°ÊãüËØóÊ≠åÊéíÁâàËäÇÂ•è"""
    if not text:
        return "ËØóÊ≠åÂÜÖÂÆπ‰∏∫Á©∫"

    lines = []
    current_line = []
    char_count = 0  # ‰∏≠ÊñáÊåâ‚ÄúÂ≠óÁ¨¶‚ÄùËÆ°Êï∞Êõ¥Ëá™ÁÑ∂

    for char in text:
        current_line.append(char)
        char_count += 1

        # ÊØè8-12‰∏™Â≠óÁ¨¶Êç¢Ë°åÔºàÊ®°ÊãüËØóÊ≠åËäÇÂ•èÔºâÔºåÊàñÈÅáÂà∞Ê†áÁÇπÊó∂Âº∫Âà∂Êç¢Ë°å
        if char_count >= random.randint(8, 12) or char in ('„ÄÇ', 'Ôºå', 'ÔºÅ', 'Ôºü', 'Ôºõ', 'Ôºö'):
            lines.append("".join(current_line))
            current_line = []
            char_count = 0

    # Ë°•ÂÖÖÊúÄÂêé‰∏ÄË°åÔºàËã•ÊúâÂâ©‰ΩôÂ≠óÁ¨¶Ôºâ
    if current_line:
        lines.append("".join(current_line))

    return "\n".join(lines)

# ÂàõÂª∫Gradio‰∫íÂä®ÁïåÈù¢
def create_interface():
    with gr.Blocks(title="AIÂÜôËØóÊú∫Âô®‰∫∫") as interface:
        gr.Markdown("# üéµ AIÂÜôËØóÊú∫Âô®‰∫∫")
        gr.Markdown("ËæìÂÖ•**‰∏≠Êñá**‰∏ªÈ¢ò/ÂºÄÂ§¥ÔºåÈÄâÊã©È£éÊ†ºÔºåÁîüÊàê‰∏≠ÊñáËØóÊ≠åÔΩû")

        with gr.Row():
            with gr.Column(scale=1):
                prompt = gr.Textbox(
                    label="ËØóÊ≠å‰∏ªÈ¢òÊàñÂºÄÂ§¥",
                    placeholder="‰æãÂ¶ÇÔºöÂ§èÂ§ú„ÄÅÊïÖ‰π°ÁöÑ‰∫ë„ÄÅÊÄùÂøµ..."
                )
                style = gr.Dropdown(
                    choices=["Áé∞‰ª£ËØó", "Âè§ËØó", "Áà±ÊÉÖËØó", "Ëá™ÁÑ∂ËØó", "Âì≤ÁêÜËØó"],
                    label="ËØóÊ≠åÈ£éÊ†º",
                    value="Áé∞‰ª£ËØó"
                )
                length = gr.Slider(
                    minimum=50, maximum=300, value=100, step=10,
                    label="ËØóÊ≠åÈïøÂ∫¶ÔºàÁõ∏ÂØπ‰∫éÂºÄÂ§¥ÁöÑÂ¢ûÈáèÔºâ"
                )
                creativity = gr.Slider(
                    minimum=0.3, maximum=1.2, value=0.7, step=0.1,
                    label="ÂàõÈÄ†ÊÄßÔºàÂÄºË∂äÈ´òÔºåËØóÂè•Ë∂äÂ§©È©¨Ë°åÁ©∫Ôºâ"
                )
                generate_btn = gr.Button("ÁîüÊàêËØóÊ≠å", variant="primary")

            with gr.Column(scale=2):
                output = gr.Textbox(label="ÁîüÊàêÁöÑËØóÊ≠å", lines=15)

        # ÁªëÂÆöÁîüÊàêÈÄªËæë
        generate_btn.click(
            fn=generate_poem,
            inputs=[prompt, style, length, creativity],
            outputs=output
        )

        # Á§∫‰æãÊ°à‰æãÔºà‰∏ÄÈîÆÁîüÊàêÂèÇËÄÉÔºâ
        gr.Examples(
            examples=[
                ["ÊúàÂÖâ‰∏ãÁöÑÊπñÈù¢", "Áé∞‰ª£ËØó", 120, 0.8],
                ["ÊÄùÂøµËøúÊñπÁöÑ‰Ω†", "Áà±ÊÉÖËØó", 100, 0.7],
                ["Â±±Èó¥Ê∏ÖÊ≥âÊµÅËøáÁü≥Áºù", "Ëá™ÁÑ∂ËØó", 150, 0.6],
                ["‰∫∫ÁîüÂ¶ÇÈÄÜÊóÖ", "Âì≤ÁêÜËØó", 130, 0.9]
            ],
            inputs=[prompt, style, length, creativity],
            outputs=output
        )

    return interface

# ÂêØÂä®ÁïåÈù¢Ôºàshare=TrueÂèØÁîüÊàêÂÖ¨ÁΩëÈìæÊé•ÔºåColabÂ§ñ‰πüËÉΩËÆøÈóÆÔºâ
if __name__ == "__main__":
    print("ÂÜôËØóÊú∫Âô®‰∫∫ÂêØÂä®‰∏≠...")
    interface = create_interface()
    interface.launch(share=True)
```
